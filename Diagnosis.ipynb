## Training and Test Data split
### Splitting the entire data set into a training and test set, using training set to fit different classifiers and correspondingly test data for prediction and evaluation.

```python
train_data, test_data = data.random_split(0.8)
```

### Logistic Regression : Predicts the probability and margins instead of actual labels themselves.

#### Fitting the model using test data and default values for L1 and L2 regularization:
##### Case 1: Using a single feature model

```python
model = graphlab.logistic_classifier.create(train_data, target='diagnosis',
                                    features = ['radius_mean'])
```

```python
single_feature_result = model.evaluate(test_data)
```

```python
single_feature_result
```

```python
model.show(view="Summary")
```

```python
## evaluating Precision, Recall, Accuracy, f1-score and AUC
```

```python
model.show(view="Evaluation")
```

### We observe a accuracy increase from a single feature model to 10-feature model, without any tradeoffs for now relating to precision and recall. Let's observe the same for a 33-featured model. 

```python
## list of 33 numeric features:
features = ['radius_mean', 'area_mean', 'texture_mean', 'perimeter_mean', 'smoothness_mean', 'concavity_mean', 'compactness_mean', 'concave_points_mean', 'symmetry_mean', 'fractal_dimensions_mean','radius_se', 'area_se', 'texture_se', 'perimeter_se', 'smoothness_se', 'concavity_se', 'compactness_se', 'concave_points_se', 'symmetry_se', 'fractal_dimension_se','radius_worst', 'area_worst', 'texture_worst', 'perimeter_worst', 'smoothness_worst', 'concavity_worst', 'compactness_worst', 'concave_points_worst', 'symmetry_worst', 'fractal_dimensions_worst']
```

```python
model_33 = graphlab.logistic_classifier.create(train_data, target='diagnosis',
                                               features = features, feature_rescaling=True)
```

```python
result_33 = model_33.evaluate(test_data)
```

```python
predictions_33 = model_33.predict(test_data)
```

```python
model_33.show(view="Summary")
```

```python
result_33
```

```python
model_33.show(view="Evaluation")
```

#### Observation: 
#### We see that on increasing the number of features 3 folds, the accuracy increases, the precision increases, but there is a decline in the recall of the model. This is knowwn as the precision recall trade-off. Here we have to choose a model in such a wy that the there is a perfect balance between the too, i.e not too low of a recall and not too low of a precision. Furthermore, increasing features does not neccesarily means that a model would be better.

### Support Vector Machine: 

```python
svm_model = graphlab.svm_classifier.create(train_data, target='diagnosis',
                                    features = features, feature_rescaling=True)
```

```python
svm_result = svm_model.evaluate(test_data)
```

```python
svm_prediction = svm_model.predict(test_data)
```

```python
svm_model.show(view="Summary")
```

```python
svm_model.show(view="Evaluation")
```

Observation: If the data is linearly separable in the input space, then LR is usually preferred as it outputs probabilities instead of hard labels and you can fine tune your performance by plotting the ROC curve and figuring out the right threshold.

The natural advantage that SVMs have over LRs is the non-linearity obtained via the use of non-linear kernels. If we compare logistic regression with SVMs with non-linear kernels, then SVMs beat LRs hands down. If the data is linearly separable in the input space, then LRs give performance comparable to SVMs, but if the data is non-linearly separable, then LRs gradually worsen depending on how bad the non-linearity is in the data and SVMs win out.

#### Decision trees tend to perform better/similar than the best model so far ( i.e logistic regression). In such as case, we use 10-fold cross validation to seperate this classifiers.

### Implementing Cross validation

```python
folds = graphlab.cross_validation.KFold(data, 5)
```

```python
params = {'target': 'diagnosis', 'max_depth': 5}
job = graphlab.cross_validation.cross_val_score(folds,
                                          graphlab.decision_tree_classifier.create,
                                          params)
print job.get_results()
```

```python
##for logistic regression
params = {'target': 'diagnosis'}
job_2 = graphlab.cross_validation.cross_val_score(folds,
                                          graphlab.logistic_classifier.create,
                                          params)
print job_2.get_results()
```

#### Observation: We see that decision tree performs better on validation set as compared to Logistic Regression, because it handles non-linear features well as compared to logistic regression. Furthermore, there is a con that decision tree is highly biased to training set and does not give as a ranked result. For now, decision tree is our model to go for diagnosis.
