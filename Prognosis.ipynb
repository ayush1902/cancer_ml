```python
import graphlab
```

```python
data = graphlab.SFrame.read_csv("wpbc.csv")
```

```python
data.head()
```

```python
graphlab.canvas.set_target('ipynb') #in order to view the visualizations in the notebook itself.
```

```python
data['outcome'] = data['outcome'].apply(lambda x:1 if x == "R" else 0)
## Converting the strng label of M and B to 1 and 0 for plotting purposes.

```

```python
#1 = Recurring, 0 = Non-recurring
data.show(view="Summary")
```

```python
print data[data['outcome']==0].num_rows()
print data[data['outcome']==1].num_rows()
```

```python
data.show(view='Bar Chart', x='radius_mean', y='outcome')
```

```python
data.show(view="Scatter Plot", x='area_mean', y='outcome')
```

```python
data.show(view="Heat Map", x="texture_mean", y="outcome")
```

```python
data.show(view="BoxWhisker Plot", x="symmetry_mean", y="diagnosis")
```

```python
train_data, test_data = data.random_split(0.8)

```

```python
features = ['time', 'radius_mean', 'area_mean', 'texture_mean', 'perimeter_mean', 'smoothness_mean', 'concavity_mean', 'compactness_mean', 'concave_points_mean', 'symmetry_mean', 'fractal_dimension_mean','radius_se', 'area_se', 'texture_se', 'perimter_se', 'smoothness_se', 'concavity_se', 'compactness_se', 'concave_points_se', 'symmetry_se', 'fractal_dimension_se','radius_worst', 'area_worst', 'texture_worst', 'perimeter_worst', 'smoothness_worst', 'concavity_worst', 'compactness_worst', 'concave_points_worst', 'symmetry_worst', 'fractal_dimension_worst', 'tumor_size', 'lymph_node_status']
```

```python
logistic_classifier = graphlab.logistic_classifier.create(train_data, target='outcome',
                                               features = features, feature_rescaling=True)
```

```python
logistic_results = logistic_classifier.evaluate(test_data)
```

```python
logistic_predictions = logistic_classifier.predict(test_data)
```

```python
logistic_results
```

```python
logistic_classifier.show(view="Summary")
```

```python
logistic_classifier.show(view="Evaluation")
```

Observation: Logistic regression performs mediocrely in case of non-linear attributes and with the added contraint of lack data entries in the data set. As opposed, the nearest neighbor is predicted to perform even worse, due to lack of number of rows.

```python
knn = graphlab.nearest_neighbor_classifier.create(train_data, target='outcome', distance='euclidean',
                                                  features=features)
```

```python
knn_results = knn.evaluate(test_data)
```

```python
knn_predictions = knn.predict(test_data)
```

```python
knn.show(view="Summary")
```

```python
knn_results
```

```python
knn_predictions
```

```python
svm_model = graphlab.svm_classifier.create(train_data, target='outcome',
                                    features = features, feature_rescaling=True)
```

```python
svm_result = svm_model.evaluate(test_data)
```

```python
svm_prediction = svm_model.predict(test_data)
```

```python
svm_model.show(view="Evaluation")
```

```python
svm_model.show(view='Summary')
```

```python
dt_model = graphlab.decision_tree_classifier.create(train_data, target='outcome', features=features, metric='auto')
```

```python
dt_result = dt_model.evaluate(test_data)
```

```python
dt_prediction = dt_model.predict(test_data)
```

```python
dt_result
```

Result: Logistic Regression performs the best among the above classifiers for a small data set and linear data.
